```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
mouse_drug_data.count()
```

```python
mouse_drug_data.head()
```

```python
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
clinical_trial_data.count()
```

```python
clinical_trial_data.head()
```

```python
# Combine the data into a single dataset
merged_data = pd.merge(clinical_trial_data,mouse_drug_data,on="Mouse ID")
merged_data.count()
```

```python
# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df_mean = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean())
grouped_df_mean.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_mean = grouped_df_mean.reset_index()
grouped_df_mean.head()
```

```python
#sample
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_df_sem = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem())
grouped_df_sem.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_sem = grouped_df_sem.reset_index()
grouped_df_sem.head()
```

```python
#sample
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_mean = pd.pivot_table(grouped_df_mean, 
                       values='Tumor Volume (mm3)', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
table_mean
```

```python
#Samples
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_sem = pd.pivot_table(grouped_df_sem, 
                       values='Tumor Volume (mm3)', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
# Preview that Reformatting worked
table_sem.head()
```

```python
# Generate the Plot (with Error Bars)
# fig, ax = plt.subplots()
# symbols ref "o","^","s","d",
#first line
x_axis = table_mean.index
plt.plot(x_axis,
         table_mean['Capomulin'],
         markevery=1,
         marker="o",
         color='r',
         label='Capomulin',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#second line
plt.plot(x_axis,
         table_mean['Infubinol'],
         markevery=1,
         marker="^",
         color='b',
         label='Infubinol',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#third line
plt.plot(x_axis,
         table_mean['Ketapril'],
         markevery=1,
         marker="s",
         color='g',
         label='Ketapril',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#fourth line
plt.plot(x_axis,
         table_mean['Placebo'],
         markevery=1,
         marker="d",
         color='black',
         label='Placebo',
         dashes=[1,1,1,1],
         alpha=0.8
        )

#errobar plot for each line
plt.errorbar(
    x_axis,
    table_mean['Capomulin'],
    yerr=table_sem['Capomulin'],
    fmt='',
    color='r',
    alpha=0.5,
    linewidth=1
)
#second line
plt.errorbar(
    x_axis,
    table_mean['Infubinol'],
    yerr=table_sem['Infubinol'],
    fmt='',
    color='b',
    alpha=0.5,
    linewidth=1
)
#third line
plt.errorbar(
    x_axis,
    table_mean['Ketapril'],
    yerr=table_sem['Ketapril'],
    fmt='',
    color='g',
    alpha=0.5,
    linewidth=1
)
#fourth line
plt.errorbar(
    x_axis,
    table_mean['Placebo'],
    yerr=table_sem['Placebo'],
    fmt='',
    color='black',
    alpha=0.5,
    linewidth=1
)

#get hline
for x in range(35,75,5):
    plt.hlines(x,min(x_axis)-5,max(x_axis)+5,alpha=0.5,linewidth=0.5)
# plt.tight_layout()

#set xlim and ylim
plt.xlim(min(x_axis)-2.5, max(x_axis)+2.5)
plt.ylim(min(table_mean['Capomulin']-2.5), max(table_mean['Ketapril'])+2.5)

#generate axis labels, title, legends
plt.legend(loc="best")

plt.title("Tumor Response To Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

#Save the Figure
plt.savefig('Images/1_tumor_response_to_treatment.png')

# Show the Figure
plt.show()
```

Obseravations:

Capomulin is the only drug that has resulted in a smaller tumor volume after 45 Days. In addition, the small stanadard error mean for Capomulin indicates the results from the samples is  likely to be representative of the population.

Infubinol and Ketapril did not seems to have any impact on reducing the tumor. The results are also in-line with the Placebo which cofirms that Infuviinol and Ketapril might not have any signifcant impact on the tumor size.

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merged_data.head()
```

```python
grouped_df_mice_count = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Mouse ID'].nunique())
grouped_df_mice_count.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_mice_count = grouped_df_mice_count.reset_index()
grouped_df_mice_count.head()
```

```python
#sample
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame
table_mice_count = pd.pivot_table(grouped_df_mice_count, 
                       values='Mouse ID', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
# Preview that Reformatting worked
table_mice_count.head()
```

```python
#sample
```

```python
# convert columns to % survived
table_mice_count['Capomulin_%'] = (table_mice_count['Capomulin']/25)*100
table_mice_count['Infubinol_%'] = (table_mice_count['Infubinol']/25)*100
table_mice_count['Ketapril_%'] = (table_mice_count['Ketapril']/25)*100
table_mice_count['Placebo_%'] = (table_mice_count['Placebo']/25)*100
table_mice_count
```

```python
# Generate the Plot (with Error Bars)
#first line
x_axis_mice = table_mice_count.index
plt.plot(x_axis_mice,
         table_mice_count['Capomulin_%'],
         markevery=1,
         marker="o",
         color='r',
         label='Capomulin',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#second line
plt.plot(x_axis_mice,
         table_mice_count['Infubinol_%'],
         markevery=1,
         marker="^",
         color='b',
         label='Infubinol',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#third line
plt.plot(x_axis_mice,
         table_mice_count['Ketapril_%'],
         markevery=1,
         marker="s",
         color='g',
         label='Ketapril',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#fourth line
plt.plot(x_axis_mice,
         table_mice_count['Placebo_%'],
         markevery=1,
         marker="d",
         color='black',
         label='Placebo',
         dashes=[1,1,1,1],
         alpha=0.8
        )


#get hline
for x in np.arange(40,110,10):
    plt.hlines(x,min(x_axis_mice)-5,max(x_axis_mice)+5,alpha=0.5,linewidth=0.5)
    
#get vline
for y in np.arange(0,50,10):
    plt.vlines(y,min(table_mice_count['Infubinol_%'])-10,max(table_mice_count['Infubinol_%'])+10,alpha=0.5,linewidth=0.5)
# plt.tight_layout()

#set xlim and ylim
plt.xlim(min(x_axis_mice)-2.5, max(x_axis_mice)+2.5)
plt.ylim(min(table_mice_count['Infubinol_%']-2.5), max(table_mice_count['Infubinol_%'])+2.5)

#generate axis labels, title, legends
plt.legend(loc="best")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%)")

#Save the Figure
plt.savefig('Images/3_survival_rates.png')

# Show the Figure
plt.show()
```

Obseravations:

Capomulin has the highest survival rate throughout and after the entire 45 days of treatment, which is at 84% by the end of the treatment. On the other hand, Infubinol and Ketapril had much lower survival rate. Infubinol's survival is even worse than Placebo which is alarming since the test subject might be better off without receiving the drug.

In this case, Capomulin clearly shows an improvement in survival rate as compared to Placebo, as well as the other two drugs.

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Tumor volume day 45 and size



# Display the data to confirm
```

```python
table_mean_t = table_mean.T
table_mean_t
```

```python
table_mean_t['%_change']=((table_mean_t[45]-table_mean_t[0])/table_mean_t[0])*100
table_mean_t
```

```python
table_mean_t['%_change']
```

```python
# Store all Relevant Percent Changes into a Tuple
```

```python
table_mean_t_selected = table_mean_t.loc[['Capomulin',
                  'Infubinol',
                  'Ketapril',
                  'Placebo'
                 ],
                 '%_change'
                ]
table_mean_t_selected
```

```python
selected_tuple=tuple(table_mean_t_selected.reset_index()['%_change'])
selected_tuple
```

```python
# Splice the data between passing and failing drugs
selected_tuple_passing=selected_tuple[0]
selected_tuple_passing
```

```python
selected_tuple_failing=selected_tuple[1:4]
selected_tuple_failing
```

```python
##### plot the bar chart
# Orient widths. Add labels, tick marks, etc
x_axis_summary = range(len(selected_tuple))

plt.bar(x_axis_summary[0],
        selected_tuple_passing,
        color='g',
        width=1,
        label='Capomulin'
       )
plt.bar(x_axis_summary[1:4],
        selected_tuple_failing,
        color='r',
        width=1
       )
#set xlim and ylim
plt.xlim(min(x_axis_summary)-0.7, max(x_axis_summary)+0.7)
plt.ylim(min(selected_tuple)-10, max(selected_tuple)+10)

#hline and vline
for x in np.arange(-20,70,20):
    plt.hlines(x,min(x_axis_summary)-1.5,max(x_axis_summary)+1.5,alpha=0.7,linewidth=0.5)
for y in np.arange(0.5,4,1):
    plt.vlines(y,selected_tuple_passing-10,max(selected_tuple_failing)+10,alpha=0.7,linewidth=0.5)
    
# #tick labels
tick_labels = table_mean_t_selected.index
tick_locations = [value+0.5 for value in x_axis_summary]
plt.xticks(tick_locations, tick_labels)

#generate axis labels, title, legends
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change)")

plt.annotate('-19%', xy=(0.5, 0), 
            xytext=(-0.1, -6),
            color='w'
            )
plt.annotate('46%', xy=(0.5, 0), 
            xytext=(0.9, 5),
            color='w'
            )
plt.annotate('57%', xy=(0.5, 0), 
            xytext=(1.9, 5),
            color='w'
            )
plt.annotate('51%', xy=(0.5, 0), 
            xytext=(2.9, 5),
            color='w'
            )
#Save the Figure
plt.savefig('Images/4_summary_bar.png')

# Show the Figure
plt.show()
```

Conclusions:

In summary, Capomulin has shown a 19% reduction in tumor volume after 45 days of treatment. In comparison, Infubinol has shown 46% increase in size, Ketapril has shown 57% increase, where as Placebo has shown 51% in size. 

At the same time, Capomulin has shown the lowest Metastatic spread as well as highest survival rate amongst all the drugs tested. In addition, with low standard error of means, the results of Capomulin on the samples are likeley to be representative of the population. It has shown an improvement over Placebo as well as the two other drugs selected for analysis. Hence, Capomulin is likely to be the best option amongst these 3 drugs in combating tumor.
