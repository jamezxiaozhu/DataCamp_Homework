```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
mouse_drug_data.count()
```

```python
mouse_drug_data.head()
```

```python
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
clinical_trial_data.count()
```

```python
clinical_trial_data.head()
```

```python
# Combine the data into a single dataset
merged_data = pd.merge(clinical_trial_data,mouse_drug_data,on="Mouse ID")
merged_data.count()
```

```python
# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df_mean = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean())
grouped_df_mean.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_mean = grouped_df_mean.reset_index()
grouped_df_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_df_sem = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem())
grouped_df_sem.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_sem = grouped_df_sem.reset_index()
grouped_df_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_mean = pd.pivot_table(grouped_df_mean, 
                       values='Tumor Volume (mm3)', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
table_mean['Capomulin']
```

```python
#Samples
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_sem = pd.pivot_table(grouped_df_sem, 
                       values='Tumor Volume (mm3)', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
# Preview that Reformatting worked
table_sem.head()
```

```python
# Generate the Plot (with Error Bars)
# fig, ax = plt.subplots()
# symbols ref "o","^","s","d",
#first line
x_axis = table_mean.index
plt.plot(x_axis,
         table_mean['Capomulin'],
         markevery=1,
         marker="o",
         color='r',
         label='Capomulin',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#second line
plt.plot(x_axis,
         table_mean['Infubinol'],
         markevery=1,
         marker="^",
         color='b',
         label='Infubinol',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#third line
plt.plot(x_axis,
         table_mean['Ketapril'],
         markevery=1,
         marker="s",
         color='g',
         label='Ketapril',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#fourth line
plt.plot(x_axis,
         table_mean['Placebo'],
         markevery=1,
         marker="d",
         color='black',
         label='Placebo',
         dashes=[1,1,1,1],
         alpha=0.8
        )

#errobar plot for each line
plt.errorbar(
    x_axis,
    table_mean['Capomulin'],
    yerr=table_sem['Capomulin'],
    fmt='',
    color='r',
    alpha=0.5,
    linewidth=1
)
#second line
plt.errorbar(
    x_axis,
    table_mean['Infubinol'],
    yerr=table_sem['Infubinol'],
    fmt='',
    color='b',
    alpha=0.5,
    linewidth=1
)
#third line
plt.errorbar(
    x_axis,
    table_mean['Ketapril'],
    yerr=table_sem['Ketapril'],
    fmt='',
    color='g',
    alpha=0.5,
    linewidth=1
)
#fourth line
plt.errorbar(
    x_axis,
    table_mean['Placebo'],
    yerr=table_sem['Placebo'],
    fmt='',
    color='black',
    alpha=0.5,
    linewidth=1
)

#get hline
for x in range(35,75,5):
    plt.hlines(x,min(x_axis)-5,max(x_axis)+5,alpha=0.5,linewidth=0.5)
# plt.tight_layout()

#set xlim and ylim
plt.xlim(min(x_axis)-2.5, max(x_axis)+2.5)
plt.ylim(min(table_mean['Capomulin']-2.5), max(table_mean['Ketapril'])+2.5)

#generate axis labels, title, legends
plt.legend(loc="best")

plt.title("Tumor Response To Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

#Save the Figure
plt.savefig('Images/1_tumor_response_to_treatment.png')
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merged_data.head()
```

```python
grouped_df_mice_count = pd.DataFrame(merged_data.groupby(['Drug','Timepoint'])['Mouse ID'].nunique())
grouped_df_mice_count.head()
```

```python
# Convert to DataFrame
# Preview DataFrame
grouped_df_mice_count = grouped_df_mice_count.reset_index()
grouped_df_mice_count.head()
```

```python
#sample
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
```

```python
table_mice_count = pd.pivot_table(grouped_df_mice_count, 
                       values='Mouse ID', 
                       index=['Timepoint'], 
                       columns=['Drug'], 
                       aggfunc=np.mean
                      )
# Preview that Reformatting worked
table_mice_count.head()
```

```python
#sample
```

```python
# convert columns to % survived
table_mice_count['Capomulin_%'] = (table_mice_count['Capomulin']/25)*100
table_mice_count['Infubinol_%'] = (table_mice_count['Infubinol']/25)*100
table_mice_count['Ketapril_%'] = (table_mice_count['Ketapril']/25)*100
table_mice_count['Placebo_%'] = (table_mice_count['Placebo']/25)*100
table_mice_count.head()
```

```python
# Generate the Plot (with Error Bars)
#first line
x_axis_mice = table_mice_count.index
plt.plot(x_axis_mice,
         table_mice_count['Capomulin_%'],
         markevery=1,
         marker="o",
         color='r',
         label='Capomulin',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#second line
plt.plot(x_axis_mice,
         table_mice_count['Infubinol_%'],
         markevery=1,
         marker="^",
         color='b',
         label='Infubinol',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#third line
plt.plot(x_axis_mice,
         table_mice_count['Ketapril_%'],
         markevery=1,
         marker="s",
         color='g',
         label='Ketapril',
         dashes=[1,1,1,1],
         alpha=0.8
        )
#fourth line
plt.plot(x_axis_mice,
         table_mice_count['Placebo_%'],
         markevery=1,
         marker="d",
         color='black',
         label='Placebo',
         dashes=[1,1,1,1],
         alpha=0.8
        )


#get hline
for x in np.arange(100,35,5):
    plt.hlines(x,min(x_axis_mice)-5,max(x_axis_mice)+5,alpha=0.5,linewidth=0.5)
# plt.tight_layout()

#set xlim and ylim
plt.xlim(min(x_axis_mice)-2.5, max(x_axis_mice)+2.5)
# plt.ylim(min(table_mice_count['Capomulin']-0.5), max(table_mice_count['Ketapril'])+0.5)

#generate axis labels, title, legends
plt.legend(loc="best")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rate(%)")

#Save the Figure
# plt.savefig('Images/3_survival_rates.png')
```
